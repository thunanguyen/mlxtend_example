```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.6, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import apriori

%timeit -n 100 -r 10 apriori(df, min_support=0.6)
```

```python
%timeit -n 100 -r 10 fpgrowth(df, min_support=0.6)
```

```python
import pandas as pd

data = pd.read_csv("BreadBasket_DMS.csv")
data
```

```python
data.drop(['Date','Time'], axis=1)
```

```python
proccessed_data = []

for i in range(data.shape[0]):
    if i == 0 or int(data['Transaction'][i]) > int(data['Transaction'][i-1]):
        proccessed_data.append([])
        
    proccessed_data[-1].append(data['Item'][i])

proccessed_data
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(proccessed_data).transform(proccessed_data)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.1, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import apriori

%timeit -r 10 apriori(df, min_support=0.1)
```

```python
from mlxtend.frequent_patterns import fpgrowth

%timeit -r 10 fpgrowth(df, min_support=0.1)
```
